```python
dict={}

def span_search(doc_id):
    with open(f"/content/drive/MyDrive/PhD/Projects/Causal Law LLM/Annotated_150/{doc_id}","r") as f2:
        content = f2.read();
    result = " ".join(line.strip() for line in content.splitlines())
    soup = BeautifulSoup(result,'html.parser')
    tags = [tag.name for tag in soup.find_all()]
    tag_text = {}
    tag_text =defaultdict(list)
    for tag in tags:
        for a in soup.find_all(tag):
            #tag_text[a.text].append(tag)
            tag_text[tag]=a.text
    dict[doc_id]=tag_text

def main():
    dir="/content/drive/MyDrive/PhD/Projects/Causal Law LLM/Annotated_150/"
    files=os.listdir(dir)
    for f in files:
        #print(f)
        span_search(f)

if __name__ == "__main__":
    main()
```

```python
# file_names = []
# for i in dict.keys():
#   file_names.append(i)

tags = []

dict['1.1958.INSC.124.txt'].keys()

for f in dict.keys():
  for t in dict[f].keys():
    if t not in tags:
      tags.append(t)

print(tags)
```

```python
dict[list(dict.keys())[0]].keys()
```

```python
# Creating Table
columns_list = tags
columns_list.append("File_Name")
df = pd.DataFrame(columns = columns_list)


for f in dict.keys():   # Getting the file names
  current_tag = []
  for t in dict[f].keys():
    current_tag.append(t)

  row = {tag: (1 if tag in current_tag else 0) for tag in tags}
  row['File_Name'] = f
  print(row)
  df.loc[len(df)] = row

```
